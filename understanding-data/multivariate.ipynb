# Scatterplot (Numerical - Numerical) (Bivariate)

```python
sns.scatterplot(x="total_bill",
                    y="tip",
                    hue="sex",
                style='smoker',
                    data=tips)
```

# Barplot (Numerical - Categorical)

```python
titanic.head()
```

```python
# in barplot we usually have categorical values and on y axis we have numerical values
# this is bivariate analysis
# sns.barplot(x='Pclass', y='Fare', data=titanic)

# the black line on top of bar indicates 'confidence interval'

# =====================================================
# this is multivariate analysis
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

# Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

# Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist=False)
```

# HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100).plot(kind='bar')
```

# Clustermap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
